```python
import numpy as np
import pandas as pd

credits_df=pd.read_csv('/kaggle/input/movie-recommendation/credits.csv')
movies_df=pd.read_csv('/kaggle/input/movie-recommendation/movies.csv')
credits_df
```

```python
movies_df
```

```python
movies_df=movies_df.merge(credits_df,on='title')
movies_df
```

```python
movies_df.info()
```

```python
movies_df=movies_df[['movie_id','title','overview','genres','keywords','cast','crew']]
movies_df.head()
```

```python
#finding no of missing values
movies_df.isnull().sum()
```

```python
#droppin missing values
movies_df.dropna(inplace= True)
```

```python
#finding duplicate values
movies_df.duplicated().sum()
```

```python
movies_df.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
        return L
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
movies_df.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i ['job']=='Director':
            L.append(i['name'])
        return L
    
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_director)
movies_df.head()
```

```python
movies_df['overview'][0]
```

```python
#separates into single separate words
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
movies_df
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies_df
```

```python
#tagging everything together
movies_df['tags']=movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
```

```python
movies_df['tags']
```

```python
movies_df
```

```python
#creating new df
new_df=movies_df[['movie_id','title','tags']]
new_df
```

```python
#removing array brackets
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x) if isinstance(x, (list, tuple)) else x)
new_df
```

```python
#lower casing
new_df['tags'] = new_df['tags'].apply(lambda x: str(x).lower())
new_df
```

```python
#feature extraction
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features= 5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors[0]
```

```python
feature_names = cv.get_feature_names_out()
print(len(feature_names))
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse= True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse= True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Fifty Shades of Grey')
```
